```python
# ROI & Sensitivity Analysis Model
Synthetic example demonstrating ROI calculation and sensitivity testing for commercial initiatives.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Initiative': ['Project A', 'Project B', 'Project C', 'Project D', 'Project E'],
    'Investment_Cost': [50000, 75000, 100000, 60000, 80000],
    'Expected_Revenue': [70000, 110000, 140000, 85000, 120000]
}
df = pd.DataFrame(data)

df['ROI'] = (df['Expected_Revenue'] - df['Investment_Cost']) / df['Investment_Cost'] * 100
print("Base ROI Calculations:")
df
```

```python
sensitivities = []
for factor in [0.8, 0.9, 1.1, 1.2]:
    temp_df = df.copy()
    temp_df['Adjusted_Cost'] = temp_df['Investment_Cost'] * factor
    temp_df['Adjusted_ROI'] = (temp_df['Expected_Revenue'] - temp_df['Adjusted_Cost']) / temp_df['Adjusted_Cost'] * 100
    sensitivities.append(temp_df['Adjusted_ROI'].mean())

factors = ['-20%', '-10%', '+10%', '+20%']
plt.bar(factors, sensitivities)
plt.title('Sensitivity Analysis: Average ROI Impact from Cost Variations')
plt.ylabel('Average ROI (%)')
plt.show()
```

```python
Business Impact: Identifies high-ROI levers for resource prioritization via trade-off testing.
```
