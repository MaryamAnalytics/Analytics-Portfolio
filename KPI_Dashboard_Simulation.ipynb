# KPI Dashboard Simulation
Illustrative KPI framework for tracking growth projects using dummy data.


```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Leads': [1000, 1200, 1100, 1300, 1400],
    'Conversions': [200, 250, 220, 280, 300],
    'Revenue': [50000, 62000, 55000, 70000, 75000]
}
df = pd.DataFrame(data)

df['Conversion_Rate'] = df['Conversions'] / df['Leads'] * 100
df['Revenue_Growth'] = df['Revenue'].pct_change() * 100
print("KPI Dashboard:")
df
```

```python
plt.plot(df['Month'], df['Conversion_Rate'], marker='o', label='Conversion Rate (%)')
plt.title('KPI Tracking: Conversion Rate Over Time')
plt.ylabel('Rate (%)')
plt.xlabel('Month')
plt.legend()
plt.show()
```

Business Impact: Enables KPI monitoring and iteration for performance improvement.
